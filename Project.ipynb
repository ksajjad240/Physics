```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
```

### Part 1: A Linear Scale
<img src="Part1.PNG" align="left">

```python
distanceList = [
    ('Mercury',          5.79*pow(10,10)),
    ('Venus',            1.08*pow(10,11)),
    ('Earth',            1.50*pow(10,11)),
    ('Mars',             2.28*pow(10,11)), 
    ('Jupiter',          7.78*pow(10,11)),
    ('Saturn',           1.43*pow(10,12)),
    ('Uranus',           2.87*pow(10,12)),
    ('Neptune',          4.50*pow(10,12)),
    ('Pluto',            5.92*pow(10,12)), 
    ('Alpha Centauri',   4.16*pow(10,16)),
    ('Barnard\'s Star',  5.68*pow(10,16)),
    ('Wolf 359',         7.38*pow(10,16))]

df = pd.DataFrame(distanceList, columns=['Solar_Object', 'Distance'])
df['Distance_Computed'] = df.Distance/pow(10, 10)
df
```

```python
df1 = df[df.Distance_Computed < 1000]
df1
```

```python
df1.plot(kind='bar', 
        x='Solar_Object', 
        y='Distance_Computed', 
        color='blue')
plt.title('Solar Objects vs. Distance from the Sun')
plt.xlabel('Solar Objects')
plt.ylabel('Distance from the Sun X 10^10 m (linear)')
plt.show()

df.plot(kind='bar', 
        x='Solar_Object', 
        y='Distance_Computed', 
        color='red')

plt.title('Solar Objects vs. Distance from the Sun')
plt.xlabel('Solar Objects')
plt.ylabel('Distance from the Sun X 10^10 m (linear)')
plt.show()
```

### Part 2: Logarithmic Scale

<img src="Part2.PNG" align="left">

```python
df['Distance_Log10'] = np.log10(df.Distance)
df
```

```python
df.plot(kind='bar', 
        x='Solar_Object', 
        y='Distance_Log10', 
        color='green')

plt.title('Solar Objects vs. Distance from the Sun')
plt.xlabel('Solar Objects')
plt.ylabel('Distance from the Sun (log10)')
plt.show()
```
