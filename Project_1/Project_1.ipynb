### Prepare Data
In this section we will prepare the data that will be required for each part of the project.

```python
distanceList = [
    (1,  '1-1', 'Mercury',          5.79*pow(10,10)),
    (2,  '1-1', 'Venus',            1.08*pow(10,11)),
    (3,  '1-1', 'Earth',            1.50*pow(10,11)),
    (4,  '1-1', 'Mars',             2.28*pow(10,11)), 
    (5,  '1-1', 'Jupiter',          7.78*pow(10,11)),
    (6,  '1-1', 'Saturn',           1.43*pow(10,12)),
    (7,  '1-1', 'Uranus',           2.87*pow(10,12)),
    (8,  '1-1', 'Neptune',          4.50*pow(10,12)),
    (9,  '1-1', 'Pluto',            5.92*pow(10,12)), 
    (10, '1-2', 'Alpha Centauri',   4.16*pow(10,16)),
    (11, '1-2', 'Barnard\'s Star',  5.68*pow(10,16)),
    (12, '1-2', 'Wolf 359',         7.38*pow(10,16)),
    (13, '3',   'Voyager-1',        2.15*pow(10,13)),
    (14, '3',   'Voyager-2',        1.78*pow(10,13)),
    (15, '4',   'Pioneer-10-2003',  1.23*pow(10,12)),
    (16, '4',   'Pioneer-10-2009',  2.31*pow(10,12)), 
    (17, '4',   'Pioneer-10-2018',  5.79*pow(10,12)),
    (18, '4',   'Aldebaran',        6.15*pow(10,17)),
    (19, '5',   'Sirius',           8.13*pow(10,16)),
    (20, '5',   'Vega',             2.39*pow(10,17)),
    (21, '5',   'Orion Nebular',    1.51*pow(10,19)),
    (22, '5',   'Milkyway Center',  2.46*pow(10,20)),
    (23, '5',   'LMC',              1.69*pow(10,21)),
    (24, '5',   'M31',              2.74*pow(10,22)),
    (25, '5',   'M104',             4.73*pow(10,23)),
    (26, '5',   'A1689-zD1',        1.21*pow(10,26)),
    (27, '5',   'HCM 6A',           1.46*pow(10,26))
]

df = pd.DataFrame(distanceList, columns=['Object_Number', 
                                         'Part_Number', 
                                         'Solar_Object', 
                                         'Distance'])

# Function to convert to linear scale. Here we will add a new column called Distance_Computed
def addLinearScale(df):
    df['Distance_Linear'] = df.Distance/pow(10, 10)
    return df

# Function to convert to logarithmic scale
def addLog10Scale(df):
    df['Distance_Log10'] = np.log10(df.Distance)
    return df
    
# Add Linear Scale
df = addLinearScale(df)
df = addLog10Scale(df)
df
```

To answer the question how many times farther away from the sun is Alpha Centauri than the planet Mercury, we need to take the ratio of the distances.

```python
df2[(df2['Part_Number'] == '1-1') | (df2['Part_Number'] == '1-2')]
alphaCentauriDistance = df2[df2['Solar_Object'] == 'Alpha Centauri'].Distance_Log10
mercuryDistance = df2[df2['Solar_Object'] == 'Mercury'].Distance_Log10

alphaCentauriDistance
mercuryDistance
distance24 = alphaCentauriDistance/float(mercuryDistance)
distance24
```

Alpha Centauri is about 1.544141 times farther away from the Sun than the planet Mercury.

Pioneer 10 was at a distance of $1.23 * 10^{12}$ meters from the Sun. To calculate its distance in 2009, we have to convert the speed from Kilometers per second to meters per year. Once calculated, we need to multiply this with 6 to get the distance covered in 6 years.

```python
distanceTraveledPerYear = 12.24 * 60 * 60 * 24 * 365 * 1000 * 6
distanceTraveledPerYear
```

```python
aldebaranDistance = df[df['Solar_Object'] == 'Aldebaran'].Distance
timeTakenToReachAldebaran = aldebaranDistance/float(distanceTraveledPerYear)
timeTakenToReachAldebaran
```

The distance of Pioneer 10 in 2009 is approximately $2.31 x 10^{12}$ meters.
The distance of Aldebaran from the Sun is $6.15 x 10{17}$ meters in linear scale and $17.788875$ in logarithmic scale.
