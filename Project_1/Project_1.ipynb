```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
```

### Part 1: A Linear Scale
<img src="Part1.PNG" align="left">

```python
distanceList = [
    (1, 'Mercury',          5.79*pow(10,10)),
    (2, 'Venus',            1.08*pow(10,11)),
    (3, 'Earth',            1.50*pow(10,11)),
    (4, 'Mars',             2.28*pow(10,11)), 
    (5, 'Jupiter',          7.78*pow(10,11)),
    (6, 'Saturn',           1.43*pow(10,12)),
    (7, 'Uranus',           2.87*pow(10,12)),
    (8, 'Neptune',          4.50*pow(10,12)),
    (9, 'Pluto',            5.92*pow(10,12)), 
    (10, 'Alpha Centauri',   4.16*pow(10,16)),
    (11, 'Barnard\'s Star',  5.68*pow(10,16)),
    (12, 'Wolf 359',         7.38*pow(10,16))]

df = pd.DataFrame(distanceList, columns=['Object_Number', 'Solar_Object', 'Distance'])
df['Distance_Computed'] = df.Distance/pow(10, 10)
df
```

```python
df1 = df[df.Distance_Computed < 1000]
df1
```

```python
# Scatter Plot
df1.plot(kind='scatter', 
        x='Object_Number', 
        y='Distance_Computed', 
        color='blue')
plt.title('Solar Objects vs. Distance from the Sun')
plt.xlabel('Solar Objects')
plt.ylabel('Distance from the Sun X 10^10 m (linear)')
plt.show()

# Bar Graph
df1.plot(kind='bar', 
        x='Solar_Object', 
        y='Distance_Computed', 
        color='blue')
plt.title('Solar Objects vs. Distance from the Sun')
plt.xlabel('Solar Objects')
plt.ylabel('Distance from the Sun X 10^10 m (linear)')
plt.show()

# Scatter Plot
df.plot(kind='scatter', 
        x='Object_Number', 
        y='Distance_Computed', 
        color='red')

plt.title('Solar Objects vs. Distance from the Sun')
plt.xlabel('Solar Objects')
plt.ylabel('Distance from the Sun X 10^10 m (linear)')
plt.show()

# Bar Graph
df.plot(kind='bar', 
        x='Solar_Object', 
        y='Distance_Computed', 
        color='red')

plt.title('Solar Objects vs. Distance from the Sun')
plt.xlabel('Solar Objects')
plt.ylabel('Distance from the Sun X 10^10 m (linear)')
plt.show()
```

### Part 2: Logarithmic Scale

<img src="Part2.PNG" align="left">

```python
df['Distance_Log10'] = np.log10(df.Distance)
df
```

```python
# Scatter Plot
df.plot(kind='scatter', 
        x='Object_Number', 
        y='Distance_Log10', 
        color='green')

plt.title('Solar Objects vs. Distance from the Sun')
plt.xlabel('Solar Objects')
plt.ylabel('Distance from the Sun (log10)')
plt.show()

# Bar Graph
df.plot(kind='bar', 
        x='Solar_Object', 
        y='Distance_Log10', 
        color='green')

plt.title('Solar Objects vs. Distance from the Sun')
plt.xlabel('Solar Objects')
plt.ylabel('Distance from the Sun (log10)')
plt.show()
```
